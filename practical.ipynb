```python
import itertools

def morseDecode(inputStringList):
    string = '' #Empty String to append the matched keys
    for letter in inputStringList:#for everyletter in inputstring 
        for key, value in morsedict.items():#for every key and value in morsecode dictionary 
            if value == letter:#if the value in morsecode dictionary is same as the letter in inputstringlist then attach the key to empty string 
                string+= key#Attaching key to empty string
    return string
    pass

def morsePartialDecode(inputStringList):
    Dictionarytest = {}# Empty Dictionary for all valid words
    string = []# empty list to append all the matching words in 
    words_file = []
    dictionaryFile = open("Desktop/dictionary.txt",'r')
    for words in dictionaryFile.readlines():
            read = words.strip('\n')#strip(\n) helps to strip the leading and trailing spaces after every word.
            Dictionarytest[read] = 0#Appending all the words in dictionaryFile to a dictionary datastructure as keys and values are zero
    combine = list(itertools.product(".-",repeat = len(inputStringList)))
    #itertools.product() works like a cartesian product for values and returns a list of multiple product sets by replacing 'x' with '.' or '-' and repeats the process until it hits the end of the list
    for each in combine:#This step help us to match every string in the combinations list0(n^2)
        emptystring =''
        Input_String =[]
        for word in range(len(inputStringList)):
            Input_String.append(each[word]+inputStringList[word].replace('x',''))
        #String concatnation of each word in inputstringlist and the combinations of replaced x 
        #After concatnation we endedup with ".x,.x,.x..,.x" string 
        #Replacing x  with empty string in the string gives us the complete string that we match the morse code with many possibilities.
            for key, value in morsedict.items():
                if Input_String[word] == value:
                    emptystring += key
        
                    #if the value in morsecode dictionary is same as the letter in inputstringlist then attach the key to empty string 
        if len(emptystring) == len(inputStringList): 
            string.append(emptystring)
            #while string replacement of x with '.','-' in many possible combiantions , we ended up with '..-.', which does not exists in morsecode dictionary  
            #As the value doesnot exists in morsecode dictionary , we will get 4letters in string than 5 letters,
            #To avoid such problems, we are comparing the lengths of both strings and appending into a list.
    
    for word in string:
        if word.lower() in Dictionarytest.keys():
            #if the inputstring is in dictionary valid words, then append them into a list
              words_file.append(word)
    return words_file
    pass
             

class Maze:
    def __init__(self):
        import numpy as np
        self.Mazearray = np.ones((11,11),dtype=None).astype(str)#created a 2d array(10*10) using numpy with all ones 
        pass

    def addCoordinate(self,x,y,blockType):
        self.Mazearray[x,y]= str(blockType)# adding zero's to the 2d array given by addCoordinate function on x and y coordinates.
        pass


    def printMaze(self):
        """
        Print out an ascii representation of the maze.
        A * indicates a wall and a empty space indicates an open space in the maze
        """
        self.Mazearray[self.Mazearray == '1.0'] = '*'# replacing the values of ones with asterisk to represent the walls in maze
        self.Mazearray[self.Mazearray == '0'] =   ' '# replacing the zero values in maze with empty spaces to represent the open spaces
        

    def findRoute(self,x1,y1,x2,y2):
        """
        This method should find a route, traversing open spaces, from the coordinates (x1,y1) to (x2,y2)
        It should return the list of traversed coordinates followed along this route as a list of tuples (x,y),
        in the order in which the coordinates must be followed
        If no route is found, return an empty list
        """
        start ='S' 
        goal = 'G'
        marked = "v"
        coordinateslist = []
        self.Mazearray[x1][y1] = start# set the start for coordinates
        self.Mazearray[x2][y2] = goal# set the goal for maze to end 
        def path(x,y): 
            
            # created a function,This function accepts the coordinates and finds the path, if it is goal then it returns true 
            # if it is wall or already marked path then it returns false. This function searches for open spaces iteratively ,
            #if it founds nothing , returns false and goes back and starts the search to find the new path.
            
            if self.Mazearray[x][y] == goal: # if it is a goal, return true
                return True
            elif self.Mazearray[x][y] == '*': # if it is a wall, return False
                return False
            elif self.Mazearray[x][y] == marked:
                #if it is visited, return False and add the visited coordinates to list. each coordinates acts as tuple as ther are immutable.
                return False
            z = (x,y)
            coordinateslist.extend([z])
            #adding the coordinates to the list.
            self.Mazearray[x][y] = marked
            # marking the already visited coordinates
            if y > 0 and path(x,y-1) == True:#if y coordinates is greater than zero , then it moves to north for exploring
                return True
            if x+1 < len(self.Mazearray)-1 and path(x+1,y) == True:# It moves to east and check for open spaces , then returns true
                return True
            if x > 0 and path(x-1,y) == True:#It moves to south and search for walls and visited elements , if yes returns False then return true
                return True
            if y+1 < len(self.Mazearray)-1 and path(x,y+1) == True:# It moves to south and searches for goal , if it finds return True then backtracks until it finds the goal.  
                return True
                
            
            return False# returns false, so it can go back and finds the new path.
            
        path(x1,y1)
        print(coordinateslist)
        string = "\n".join(" ".join(str(el) for el in row)for row in self.Mazearray)
        print(string)
        
def morseCodeTest():
    """
    This test program passes the morse code as a list of strings for the word
    HELLO to the decode method. It should receive a string "HELLO" in return.
    This is provided as a simple test example, but by no means covers all possibilities, and you should
    fulfill the methods as described in their comments.
    """
    global morsedict
    
    morsedict ={'A':'.-','B':'-...','C':'-.-.','D':'-..','E':'.','G':'--.','H':'....','I':'..','J':'.---','k':'-.-','L':'.-..','M':'--','N':'-.','O':'---','P':'.--.','Q':'--.-','R':'.-.','S':'...','T':'-','U':'..-','V':'...-','W':'.--',
            'X':'-..-','Y':'-.--','Z':'--..','1':'.----','2':'..---','3':'...--','4':'....-','5':'.....','6':'-....','7':'--...','8':'---...','9':'----.','0':'-----'}

    
    hello = ['....','.','.-..','.-..','---']
    print(morseDecode(hello))
def partialMorseCodeTest():


    """
    This test program passes the partial morse code as a list of strings 
    to the morsePartialDecode method. This is provided as a simple test example, but by
    no means covers all possibilities, and you should fulfill the methods as described in their comments.
    """
    morsedict ={'A':'.-','B':'-...','C':'-.-.','D':'-..','E':'.','G':'--.','H':'....','I':'..','J':'.---','k':'-.-','L':'.-..','M':'--','N':'-.','O':'---','P':'.--.','Q':'--.-','R':'.-.','S':'...','T':'-','U':'..-','V':'...-','W':'.--',
            'X':'-..-','Y':'-.--','Z':'--..','1':'.----','2':'..---','3':'...--','4':'....-','5':'.....','6':'-....','7':'--...','8':'---...','9':'----.','0':'-----'}

    # This is a partial representation of the word TEST, amongst other possible combinations
    test = ['x','x','x..','x']
    print(morsePartialDecode(test))

    # This is a partial representation of the word DANCE, amongst other possible combinations
    dance = ['x..','x-','x.','x.-.','x']
    print(morsePartialDecode(dance))  
def mazeTest():
    """
    This sets the open space coordinates for the example
    maze in the assignment.
    The remainder of coordinates within the max bounds of these specified coordinates
    are assumed to be walls
    """
    
    myMaze = Maze()
    myMaze.addCoordinate(1,0,0)
    myMaze.addCoordinate(1,1,0)
    myMaze.addCoordinate(7,1,0)
    myMaze.addCoordinate(1,2,0)
    myMaze.addCoordinate(2,2,0)
    myMaze.addCoordinate(3,2,0)
    myMaze.addCoordinate(4,2,0)
    myMaze.addCoordinate(6,2,0)
    myMaze.addCoordinate(7,2,0)
    myMaze.addCoordinate(4,3,0)
    myMaze.addCoordinate(7,3,0)
    myMaze.addCoordinate(4,4,0)
    myMaze.addCoordinate(7,4,0)
    myMaze.addCoordinate(3,5,0)
    myMaze.addCoordinate(4,5,0)
    myMaze.addCoordinate(7,5,0)
    myMaze.addCoordinate(1,6,0)
    myMaze.addCoordinate(2,6,0)
    myMaze.addCoordinate(3,6,0)
    myMaze.addCoordinate(4,6,0)
    myMaze.addCoordinate(5,6,0)
    myMaze.addCoordinate(6,6,0)
    myMaze.addCoordinate(7,6,0)
    myMaze.addCoordinate(5,7,0)
    myMaze.printMaze()
    myMaze.findRoute(1,0,4,6)
    
def main():
    morseCodeTest()
    partialMorseCodeTest()
    mazeTest()
   
    
if(__name__ == "__main__"):
    main()

        
```
